```python
from faker import Faker
import csv
import pandas as pd
```

```python

fake = Faker()

# Open a CSV file for writing
with open(
    r"C:\Users\user\Desktop\Stutern Inter-track project\loan_data.csv",
    "w",
    newline="",
) as file:
    writer = csv.writer(file)
    # Write header row

    writer.writerow(
        [
            "Name",
            "Phone Number",
            "Email Address",
            "Date of Birth",
            "Address",
            "National Identity Number",
            "Employment",
            "Income per month",
            "Loan Type",
            "Repayment Type",
            "Credit Score",
            "Collateral Type",
            "Collateral Value",
            "Guarantor's Name",
            "Guarantor's Phone Number",
            "Guarantor's Email",
            "Guarantor's Date of Birth",
            "Guarantor's Address",
            "Guarantor's Social Security Number",
            "Guarantor's Relationship",
            "Guarantor's Employment",
            "Guarantor's Other sources of income",
            "Guarantor's Income per month",
            "Loan Status",
        ]
    )

    # Generate 500,000 rows of data
    for _ in range(50):
        name = fake.name()
        phone_number = (
            fake.random_element(["080", "070", "081", "090"])
            + "-"
            + fake.numerify("####")
            + "-"
            + fake.numerify("####")
        )

        email = fake.email()
        date_of_birth = fake.date_of_birth()
        address = fake.address()
        national_indentity_number = fake.random_number(digits=11)
        employment = fake.random_element(
            ["FullTime", "PartTime", "Self-Employed", "Contract"]
        )
        income_per_month = fake.random_int(min=1000, max=10000)
        loan_type = fake.random_element(
            ["Housing Loan", "Student Loan", "Agricultural Loan", "Business Loan"]
        )

        repayment_type = fake.random_element(
            [
                "Interest only",
                "Principal & Interest",
                "Capitalized Interest",
                "Housing Loan",
                "Others",
            ]
        )

        credit_score = fake.random_number(digits=3)
        collateral_type = fake.random_element(["Business Equipment", "Real Estate"])
        collateral_value = fake.random_int(min=1000, max=10000)
        guarantor_name = fake.name()
        guarantor_phone_number = fake.phone_number()
        guarantor_email = fake.email()
        guarantor_dob = fake.date_of_birth()
        guarantor_address = fake.address()
        guarantor_ssn = fake.ssn()
        guarantor_relationship = fake.random_element(
            ["Parent", "Sister", "Brother", "Spouse", "Friend", "Employer"]
        )
        guarantor_employment = fake.random_element(
            ["FullTime", "PartTime", "Retiree", "Contract", "Business Owner"]
        )

        guarantor_other_sources_income = fake.random_element(
            [
                "Rental Income",
                "Investment Income",
                "Freelancing",
                "Business Profits",
                "Royalties",
                "Agriculture and Farming",
                "Rental of Assets",
            ]
        )

        guarantor_income_per_month = fake.random_int(min=1000, max=10000)
        loan_status = fake.random_element(["Approved", "Rejected"])

        # Write the data to the CSV file
        writer.writerow(
            [
                name,
                phone_number,
                email,
                date_of_birth,
                address,
                national_indentity_number,
                employment,
                income_per_month,
                loan_type,
                repayment_type,
                credit_score,
                collateral_type,
                collateral_value,
                guarantor_name,
                guarantor_phone_number,
                guarantor_email,
                guarantor_dob,
                guarantor_address,
                guarantor_ssn,
                guarantor_relationship,
                guarantor_employment,
                guarantor_other_sources_income,
                guarantor_income_per_month,
                loan_status,
            ]
        )
```
